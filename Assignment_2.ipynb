# B) Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from Mileage, CC from Engine, bhp from Power, and lakh from New_price).

```python
fuel_density = 0.832

def convert_to_kmpl(value):
    value=str(value)
    if 'km/kg' in value:
        # Extract the numeric part and convert
        km_per_kg = float(value.replace(' km/kg', ''))
        return km_per_kg * fuel_density  # Convert to kmpl
    elif 'kmpl' in value:
        # Extract the numeric part for values already in kmpl
        return float(value.replace(' kmpl', ''))
    else:
         # Return the original value if it's not in expected format
        # or If it's already a number (e.g., float)
        try:
            return float(value)
        except ValueError:
            return None  # Handle unexpected formats if any

# Apply the conversion function to each row in the 'efficiency' column
df1['Mileage'] = df1['Mileage'].apply(convert_to_kmpl)

df1
```

```python
df1['Engine'] = df1['Engine'].astype(str).str.replace(' CC', '', regex=False).astype(float)
df1['Power'] = df1['Power'].astype(str).str.replace(' bhp', '', regex=False).astype(float)
df1
```

```python
df1.fillna(df1.median(numeric_only=True), inplace=True)
df1
```

#D)Create one more feature and add this column to the dataset (you can use mutate function in R for this). For example, you can calculate the current age of the car by subtracting Year value from the current year.

```python
from datetime import datetime
current_year = datetime.now().year

# Calculate the age of each car and add it as a new column
df1['car_age'] = current_year - df1['Year']

df1
```

#e) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or their equivalent operations in python) on this dataset.

```python
# Select specific columns (similar to SELECT in SQL)
selected_df1 = df1[['Name', 'car_age']]
print("Selected Columns:\n", selected_df1)

# Filter rows based on a condition (similar to FILTER)
filtered_df1 = df1[df1['car_age'] > 5]
print("\nFiltered Rows (car_age > 5):\n", filtered_df1)

# Rename columns (similar to R's rename or SQL's AS)
renamed_df1 = df1.rename(columns={'car_name': 'make', 'car_age': 'age_of_car'})
print("\nRenamed Columns:\n", renamed_df1)

# Mutate: Create a new column based on a condition or calculation
df1['is_new'] = df1['car_age'] < 3
print("\nMutated with New Column (is_new):\n", df1)

# Arrange: Sort by one or more columns (similar to ORDER BY)
arranged_df1 = df1.sort_values(by='car_age', ascending=False)
print("\nArranged by car_age (Descending):\n", arranged_df1)

# Summarize with Group By: Group data and calculate summary statistics
grouped_df1 = df1.groupby('Name').agg(
    avg_age=('car_age', 'mean'),
    count=('car_age', 'size')
).reset_index()
print("\nGrouped and Summarized by car_name:\n", grouped_df1)
```
